Anlise da vriavel tempo

```python
#Analisar valores invlidos do atributo "Tempo"
for value in df['Tempo']:
  try:
    float(value)
  except ValueError:
    print(f"Valor invlido: {value}")
```

```python
#Anlise da mdia do tempo das msicas
#Remover linhas com valores invlidos
df = df[df['Tempo'] != ' ' ]
#Converter os valores em string para float
df['Tempo'] = df['Tempo'].astype(float)
#Plotar o grfico
ax = sns.distplot(df.Tempo)
ax.set(xlabel = "Tempo da msica", ylabel = "Densidade")
ax.set_title ("Mdia do tempo das msicas")
```

```python
#Anlise da nota mdia dos filmes
ax = sns.boxplot(x=df.Tempo)
ax.set(xlabel ="Tempo da msica")
ax.set_title("Distribuio do tempo mdio das msicas")
```

```python
#Intervalo de confiana da mdia com t e ztest
from statsmodels.stats.weightstats import zconfint

zconfint(df.Tempo)
```

Analisado o intervalo de confiana que temos para as mdias do tempo das msicas. Isso quer dizer que as msicas tem uma mdia 121.48, que  a mdia das msicas da base de dados analisada, mas queremos criar uma generalizao, a partir dessa amostra, sobre as msicas que existem no mundo.

```python
# Incluir nova coluna no dataframe com o calculo do tempo X 2.
df['Tempo_2'] = df['Tempo'] * 2
df['Tempo_2']
```

Analisar os gneros e encontrar os mais populares

```python
df['Genre']=df['Genre'].astype(str)
df["Genre"][df["Genre"] == "[]"] = np.nan
df["Genre"] = df["Genre"].fillna(0)
#here we get rid of useless symbols to be able to separate genres
df.Genre=df.Genre.str.replace("[", "")
df.Genre=df.Genre.str.replace("]", "")
df.Genre=df.Genre.str.replace("'", "")
#now we devide genre strings by comma
df["Genre"] = df["Genre"].str.split(",")
#next command separates rows based on genres, so for each song that is marked with several genres,
#now we'll have multiple rows with one genre for each row
df=df.explode('Genre')
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.subplots()
df.Genre.value_counts()[:25].plot(ax=ax, kind = "pie")
ax.set_ylabel("")
ax.set_title("Top 25 most popular genres")
plt.show()
```

```python
#Classificar o dataframe com base no "Nmero de tempos mapeados" e agrupar por artista e somar o nmero de vezes mapeado para cada artista

df_numbercharted=df.groupby('Artist').sum().sort_values('Number of Times Charted', ascending=False)
df_numbercharted=df_numbercharted.reset_index()
df_numbercharted
```

```python
#Os 10 artistas mais populares com base em sua apario nas paradas
plt.figure(figsize=(20,8))
plt.title("Top 10 Artists with Highest Numbers of Times Charted")
sns.barplot(x='Artist',y='Number of Times Charted',data=df_numbercharted.head(10))
```

```python
# converter tipo data da fatura para data
df['Release Date'] = pd.to_datetime(df['Release Date'], dayfirst=True)
# criar novas colunas para ano_ms, ano, ms, dia da semana
df['Ano_Ms'] = df['Release Date'].dt.to_period('M')
df['Ano'] = df['Release Date'].dt.strftime('%Y')
df['Ms'] = df['Release Date'].dt.strftime('%b')
df['Dia_da_Semana'] = df['Release Date'].dt.strftime('%a')
df.head()
```

```python
#Verificar a quantidade de msicas lanadas por Ano
df.groupby(['Ano'])['Song ID'].agg(['count'])
```

```python
#Verificar a quantidade de msicas lanadas por dia da semana
df.groupby(['Dia_da_Semana'])['Song ID'].agg(['count'])
```

Observa-se que o o dia da semana "SEXTA"  o dia que tem o maior nmero de msicas lanadas.

```python
# Agrupar as vendas por dia da semana e calcular a soma
Lancamento = df.groupby(['Ano'])['Song ID'].count()

# Obter os dias da semana como ndice
Ano_Lancamento = Lancamento.index

# Obter os valores das vendas
total_musicas = Lancamento.values

plt.bar(Ano_Lancamento, total_musicas)
plt.xlabel('Ano_Lancamento', color='gray')
plt.ylabel('Qtd. msicas', color='gray')
plt.title('Quantidade de msicas por ano de  lanamento', color='blue')

# Rotacionar os rtulos do eixo x
plt.xticks(rotation=45)

plt.show()
```

Observa-se que o ano de 2020  o ano com o maior nmero de msicas lanadas.

```python
# Agrupar as vendas por dia da semana e calcular a soma
Lancamento = df.groupby(['Ms'])['Song ID'].count()

# Obter os dias da semana como ndice
Ms_Lancamento = Lancamento.index

# Obter os valores das vendas
total_musicas = Lancamento.values

plt.bar(Ms_Lancamento, total_musicas)
plt.xlabel('Ms_Lancamento', color='gray')
plt.ylabel('Qtd. msicas', color='gray')
plt.title('Quantidade de msicas por ms de  lanamento', color='blue')

# Rotacionar os rtulos do eixo x
plt.xticks(rotation=45)

plt.show()
```

Observa-se que o ms de Maio  o ms com o maior nmero de msicas lanadas.
