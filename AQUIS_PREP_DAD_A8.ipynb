Análise da váriavel tempo

```python
#Analisar valores inválidos do atributo "Tempo"
for value in df['Tempo']:
  try:
    float(value)
  except ValueError:
    print(f"Valor inválido: {value}")
```

```python
#Análise da média do tempo das músicas
#Remover linhas com valores inválidos
df = df[df['Tempo'] != ' ' ]
#Converter os valores em string para float
df['Tempo'] = df['Tempo'].astype(float)
#Plotar o gráfico
ax = sns.distplot(df.Tempo)
ax.set(xlabel = "Tempo da música", ylabel = "Densidade")
ax.set_title ("Média do tempo das músicas")
```

```python
#Análise da nota média dos filmes
ax = sns.boxplot(x=df.Tempo)
ax.set(xlabel ="Tempo da música")
ax.set_title("Distribuição do tempo médio das músicas")
```

```python
#Intervalo de confiança da média com t e ztest
from statsmodels.stats.weightstats import zconfint

zconfint(df.Tempo)
```

Analisado o intervalo de confiança que temos para as médias do tempo das músicas. Isso quer dizer que as músicas tem uma média 121.48, que é a média das músicas da base de dados analisada, mas queremos criar uma generalização, a partir dessa amostra, sobre as músicas que existem no mundo.

```python
# Incluir nova coluna no dataframe com o calculo do tempo X 2.
df['Tempo_2'] = df['Tempo'] * 2
df['Tempo_2']
```

```python
#Analisar quais músicas possuem tempo igual a 0
df.query("Tempo == 0").head()
```

Analisar os gêneros e encontrar os mais populares

```python
df['Genre']=df['Genre'].astype(str)
df["Genre"][df["Genre"] == "[]"] = np.nan
df["Genre"] = df["Genre"].fillna(0)
#here we get rid of useless symbols to be able to separate genres
df.Genre=df.Genre.str.replace("[", "")
df.Genre=df.Genre.str.replace("]", "")
df.Genre=df.Genre.str.replace("'", "")
#now we devide genre strings by comma
df["Genre"] = df["Genre"].str.split(",")
#next command separates rows based on genres, so for each song that is marked with several genres,
#now we'll have multiple rows with one genre for each row
df=df.explode('Genre')
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.subplots()
df.Genre.value_counts()[:25].plot(ax=ax, kind = "pie")
ax.set_ylabel("")
ax.set_title("Top 25 most popular genres")
plt.show()
```

```python
#Classificar o dataframe com base no "Número de tempos mapeados" e agrupar por artista e somar o número de vezes mapeado para cada artista

df_numbercharted=df.groupby('Artist').sum().sort_values('Number of Times Charted', ascending=False)
df_numbercharted=df_numbercharted.reset_index()
df_numbercharted
```

```python
#Os 10 artistas mais populares com base em sua aparição nas paradas
plt.figure(figsize=(20,8))
plt.title("Top 10 Artists with Highest Numbers of Times Charted")
sns.barplot(x='Artist',y='Number of Times Charted',data=df_numbercharted.head(10))
```

```python
# converter tipo data da fatura para data
df['Release Date'] = pd.to_datetime(df['Release Date'], dayfirst=True)
# criar novas colunas para ano_mês, ano, mês, dia da semana
df['Ano_Mês'] = df['Release Date'].dt.to_period('M')
df['Ano'] = df['Release Date'].dt.strftime('%Y')
df['Mês'] = df['Release Date'].dt.strftime('%b')
df['Dia_da_Semana'] = df['Release Date'].dt.strftime('%a')
df.head()
```

```python
#Verificar a quantidade de músicas lançadas por Ano
df.groupby(['Ano'])['Song ID'].agg(['count'])
```

```python
#Verificar a quantidade de músicas lançadas por dia da semana
df.groupby(['Dia_da_Semana'])['Song ID'].agg(['count'])
```

Observa-se que o o dia da semana "SEXTA" é o dia que tem o maior número de músicas lançadas.

```python
# Agrupar as vendas por dia da semana e calcular a soma
Lancamento = df.groupby(['Ano'])['Song ID'].count()

# Obter os dias da semana como índice
Ano_Lancamento = Lancamento.index

# Obter os valores das vendas
total_musicas = Lancamento.values

plt.bar(Ano_Lancamento, total_musicas)
plt.xlabel('Ano_Lancamento', color='gray')
plt.ylabel('Qtd. músicas', color='gray')
plt.title('Quantidade de músicas por ano de  lançamento', color='blue')

# Rotacionar os rótulos do eixo x
plt.xticks(rotation=45)

plt.show()
```

Observa-se que o ano de 2020 é o ano com o maior número de músicas lançadas.

```python
# Agrupar as vendas por dia da semana e calcular a soma
Lancamento = df.groupby(['Mês'])['Song ID'].count()

# Obter os dias da semana como índice
Mês_Lancamento = Lancamento.index

# Obter os valores das vendas
total_musicas = Lancamento.values

plt.bar(Mês_Lancamento, total_musicas)
plt.xlabel('Mês_Lancamento', color='gray')
plt.ylabel('Qtd. músicas', color='gray')
plt.title('Quantidade de músicas por mês de  lançamento', color='blue')

# Rotacionar os rótulos do eixo x
plt.xticks(rotation=45)

plt.show()
```

Observa-se que o mês de Maio é o mês com o maior número de músicas lançadas.
